## Groupby

```python
data = {
    'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
    'Person': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
    'Sales': [200, 120, 340, 124, 243, 350]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby('Company')['Sales'].agg(['sum', 'mean', 'count', 'min', 'max'])
```

```python
df.groupby('Company')['Sales'].mean()
```

```python
def coeff_variation(x):
  return x.std()/x.mean() if x.mean() !=0 else 0

custom_agg = df.groupby('Company')['Sales'].agg(['sum', coeff_variation])
custom_agg

```

```python
df.groupby('Company')['Sales'].sum(numeric_only = True)
```

```python
df.groupby('Company').describe()
```

```python
df.describe()
```

```python
df.groupby('Company').describe().T    # T is transpose which changes rows to columns
```

```python
df.groupby('Company').describe().T['FB']
```

```python
df.groupby('Company')['Sales'].sum()
```

```python
df.groupby('Company')['Sales'].transform('sum')   #transform function, takes the group statistics and gives a copy to each element in the group
```

```python
company_avg_agg = df.groupby('Company')['Sales'].mean()
company_avg_agg
```

```python
df['Company_Avg_Sales'] = df.groupby('Company')['Sales'].transform('mean')
df
```

## Merge & Join

1. Merge combines data on columns *pd.merge*
2. Joincombined based on index *df.join*

```python
employees = pd.DataFrame({
    'Employee_ID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'DeptID': [1, 1, 2, 3]
})

departments = pd.DataFrame({
    'DeptID': [1, 2, 3],
    'DeptName': ['HR', 'IT', 'Finance']
})

salaries = pd.DataFrame({
    'Employee_ID': [101, 102, 104],
    'Salary': [70000, 80000, 75000]
})
```

```python
employees
```

```python
departments
```

```python
salaries
```

```python
pd.merge(employees, departments, on ='DeptID')
```

```python
pd.merge(departments, employees, on ='DeptID')
```

```python
pd.merge(employees, salaries, on = 'Employee_ID')
```

```python
pd.merge(salaries, employees, on = 'Employee_ID')
```

```python
pd.merge(employees, salaries, on = 'Employee_ID', how ='right')
```

```python
pd.merge(employees, salaries, on = 'Employee_ID', how ='left')
```

```python
pd.merge(employees, salaries, on = 'Employee_ID', how ='inner')
```

```python
pd.merge(employees, salaries, on = 'Employee_ID', how ='outer')
```
